```python
import cv2  # importing cv
from skimage import transform
import matplotlib.pyplot as plt

def getImage(path):
    return cv2.imread(path)

def rotate_img(sample_image_path, angle):
    img = getImage(sample_image_path)
    return transform.rotate(img, angle=angle, cval=255)


rotated = rotate_img("/content/gdrive/MyDrive/Kinship_Data/train-faces/F0001/MID1/P00001_face0.jpg", -55)
original_show = plt.imshow(getImage("/content/gdrive/MyDrive/Kinship_Data/train-faces/F0001/MID1/P00001_face0.jpg"))
plt.show()
rotated_show = plt.imshow(rotated)
plt.show()

```

```python
import glob, os
from random import randrange
os.chdir("/content/gdrive/MyDrive/Kinship_Data/train-faces/F0001/MID1/")
rotated_images = []
angles = [-55, 55, -145, 145]
for file in glob.glob("*.jpg"):
    path = "/content/gdrive/MyDrive/Kinship_Data/train-faces/F0001/MID1/" + file
    plt.imshow(getImage(path))
    plt.show()
    rotated = rotate_img(path, angles[randrange(len(angles) - 1)])
    rotated_images.append(rotated)
    plt.imshow(rotated)
    plt.show()

    
```

```python
family_paths = glob.glob("/content/gdrive/MyDrive/Kinship_Data/train-faces/*")

rotated_families = {}
for family_path in family_paths:
    mid_paths = glob.glob(family_path + "/*")
    # print(mid_paths)
    rotated_mids = {}
    for mid_path in mid_paths:
        img_paths = glob.glob(mid_path + "/*")
        rotated = rotate_img(img_paths, randrange(len(angles) - 1))
        rotated_mids[mid_path] = rotated
    rotated_families[family_path] = rotated_mids


```
